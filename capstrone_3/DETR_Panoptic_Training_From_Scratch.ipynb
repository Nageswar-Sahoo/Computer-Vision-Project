```python
# COCO classes
CLASSES = ['NA','rebar', 'spall','crack']
# colors for visualization
COLORS = [[0.000, 0.447, 0.741], [0.850, 0.325, 0.098], [0.929, 0.694, 0.125],
          [0.494, 0.184, 0.556], [0.466, 0.674, 0.188], [0.301, 0.745, 0.933]]
```

```python
# standard PyTorch mean-std input image normalization
transform = T.Compose([
    T.Resize(800),
    T.ToTensor(),
    T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# for output bounding box post-processing
def box_cxcywh_to_xyxy(x):
    x_c, y_c, w, h = x.unbind(1)
    b = [(x_c - 0.5 * w), (y_c - 0.5 * h),
         (x_c + 0.5 * w), (y_c + 0.5 * h)]
    return torch.stack(b, dim=1)

def rescale_bboxes(out_bbox, size):
    img_w, img_h = size
    b = box_cxcywh_to_xyxy(out_bbox)
    b = b * torch.tensor([img_w, img_h, img_w, img_h], dtype=torch.float32)
    return b

def plot_results(pil_img, prob, boxes):
    plt.figure(figsize=(16,10))
    plt.imshow(pil_img)
    ax = plt.gca()
    for p, (xmin, ymin, xmax, ymax), c in zip(prob, boxes.tolist(), COLORS * 100):
        ax.add_patch(plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin,
                                   fill=False, color=c, linewidth=2))
        cl = p.argmax()
        print(xmin , ymin ,xmax, ymax )

        text = f'{CLASSES[cl]}: {p[cl]:0.2f}'
        ax.text(xmin, ymin, text, fontsize=15,
                bbox=dict(facecolor='yellow', alpha=0.5))
    plt.axis('off')
    plt.show()       
```

```python
def detect(im, model, transform):
    # mean-std normalize the input image (batch-size: 1)
    img = transform(im).unsqueeze(0)

    # demo model only support by default images with aspect ratio between 0.5 and 2
    # if you want to use images with an aspect ratio outside this range
    # rescale your image so that the maximum size is at most 1333 for best results
    assert img.shape[-2] <= 1600 and img.shape[-1] <= 1600, 'demo model only supports images up to 1600 pixels on each side'

    # propagate through the model
    outputs = model(img)
    #print(outputs)

    # keep only predictions with 0.7+ confidence
    probas = outputs['pred_logits'].softmax(-1)[0, :, :-1]
    #print(probas.max(-1).values)
    keep = probas.max(-1).values > 0.5

    # convert boxes from [0; 1] to image scales
    bboxes_scaled = rescale_bboxes(outputs['pred_boxes'][0, keep], im.size)
    #print(bboxes_scaled)
    return probas[keep], bboxes_scaled
```

```python
from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import pylab
pylab.rcParams['figure.figsize'] = (8.0, 10.0)  # Import Libraries
import matplotlib
#matplotlib.use('Qt5Agg')
import matplotlib.pyplot as plt
import cv2
import os
```

```python
img_id = range(1,250)
dataDir = 'coco'
mul_dataType = 'annotations/instances_test2017'
mul_annFile = '{}/{}.json'.format(dataDir, mul_dataType)
img_dir = "coco/test2017"
white = [255,255,255]
mul_coco = COCO(mul_annFile)

for id in img_id:
  try:
    imgId = mul_coco.getImgIds(imgIds=[id])
    img = mul_coco.loadImgs(imgId)[0]
    mul_annIds = mul_coco.getAnnIds(imgIds=imgId, iscrowd=None)
    mul_anns = mul_coco.loadAnns(mul_annIds)
    for region in mul_anns:

        actual = io.imread(img_dir + '/' + img['file_name'])
        annotated = io.imread(img_dir + '/' + img['file_name'])
        annotated_predicted_1 = io.imread(img_dir + '/' + img['file_name'])
        annotated_predicted = Image.open(img_dir + '/' + img['file_name'])

        scores, boxes = detect(annotated_predicted, detr, transform)
        for p, bbox , c in zip(scores, boxes.tolist(), COLORS * 100):    
          cl = p.argmax()
          text = f'{CLASSES[cl]}: {p[cl]:0.2f}'  
          #annotated_predicted = cv2.rectangle(annotated_predicted, (int(xmin),int(ymin)), (int(xmax - xmin), int(ymax - ymin)), (255, 0, 0), 2)    
          #annotated_predicted =cv2.rectangle(annotated_predicted, (int(xmin), int(ymin)), (int(xmax - xmin), int(ymax - ymin)), (255, 0, 0), 2 )
          bbox=list(map(int, bbox))
          #annotated_predicted = cv2.rectangle(annotated_predicted, (bbox[0], bbox[1]), (bbox[0] + bbox[2], bbox[1] + bbox[3]), (255, 0, 0), 2)
          color = list(np.random.random(size=3) * 256)
          annotated_predicted_1 = cv2.rectangle(annotated_predicted_1, (bbox[0], bbox[1]), (bbox[2] , bbox[3]  ),color, 2)
          cv2.putText(annotated_predicted_1, text, (bbox[0], bbox[1]-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)        
        



 

        bboxes = []
        for box in mul_anns:
           bboxes.append(box['bbox'])

        for bbox in bboxes:
            annotated = cv2.rectangle(annotated, (bbox[0], bbox[1]), (bbox[0] + bbox[2], bbox[1] + bbox[3]), (255, 0, 0), 2)
            cv2.putText(annotated, CLASSES[region['category_id']], (bbox[0], bbox[1]-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)


        
        annotated= cv2.copyMakeBorder(annotated,20,20,20,20,cv2.BORDER_CONSTANT,value=white)
        actual= cv2.copyMakeBorder(actual,20,20,20,20,cv2.BORDER_CONSTANT,value=white)
        annotated_predicted_1= cv2.copyMakeBorder(annotated_predicted_1,20,20,20,20,cv2.BORDER_CONSTANT,value=white)

        cat_image = np.concatenate([actual, annotated, annotated_predicted_1], axis=1)
        path = 'ou_img'
        cv2.imwrite(os.path.join(path , str(id) + ".jpg"), cat_image)
  except:
      print("An exception occurred")
```
